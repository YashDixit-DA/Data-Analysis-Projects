# Part D – Customer Analysis (RFM) (20 Marks)

# Using CustomerID:
## 1. Compute:

## Recency

```python

reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=30)
```

```python
recency_df = (
    df.groupby('CustomerID')['InvoiceDate']
      .max()
      .reset_index()
)

recency_df['Recency'] = (reference_date - recency_df['InvoiceDate']).dt.days
```

```python
recency_df[['CustomerID', 'Recency']].sort_values( by = "Recency")
```

```python
bins = [0,30,60,90,float('inf')]
labels = ["Very active customer", "Active","at risk","Inactive"]
```

```python
recency_df["Recency Segmenation"] = pd.cut(recency_df['Recency'],
    bins=bins,
    labels=labels,
    right=True)
```

```python
recency_df
```

## Recency based on Each Customer Average Days gap between purchase  

```python
df = df.sort_values(['CustomerID', 'InvoiceDate'])
```

```python
df['Days_Between_Purchases'] = (df.groupby("CustomerID")["InvoiceDate"].diff().dt.days)
```

```python
avg_gap_per_customer = df.groupby("CustomerID")["Days_Between_Purchases"].mean().reset_index(name='Avg_Days_Between_Purchases')
```

```python
df = df.merge(avg_gap_per_customer,on='CustomerID',how='left')
```

```python
overall_avg_days = avg_gap_per_customer['Avg_Days_Between_Purchases'].mean()
```

```python
avg_gap_per_customer['Avg_Days_Between_Purchases'] = (
    avg_gap_per_customer['Avg_Days_Between_Purchases'].fillna(0)
)
```

```python
print(f"Average days between purchases across all customers: {overall_avg_days:.2f} days")
```
